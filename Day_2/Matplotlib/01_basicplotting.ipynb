# Matplotlib Basics
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

using the pyplot submodule and numpy, we can now generate simple plots!

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

Right now this code plots the two points (of which the coordinates (0,0) and (6,250) are specified in the two arrays xpoints and ypoints), which are connected via a line. If we want to make it more clear that we are actually only plotting single points, we can specify the type of line that is produced in the plot:

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

In the code above, we have specified that we only want to see the points that we are plotting (without the line connecting them) by running:

plt.plot(xpoints, ypoints, 'o')

But what does the term 'o' actually do? The plot function can produce very customizable outcomes. Using different arguments, we can change among others the linestyle of our lines, the marker that marks the points or the markersize we want.

All of these arguments are specified in the documentation of matplotlib: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

There are many other ways in which plots can be customized. We can add legends, titles, axis names and much more.
