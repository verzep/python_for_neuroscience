# Artists

Now that you have learned about the basic plotting functions of matplotlib, we will move on to more advanced ways to think about plots.

In matplotlib whenever we plot something, we create a "Figure". You can think of a figure as the empty canvas that you draw your plots onto. Onto a Figure, we can draw plots using the "Axes" functionality. Each "Ax" that we put into a Figure can have different characteristics.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.random.normal(-20, 15, 500) # this code creates numbers in a normal distribution with the mean = -20, the standard deviation = 15 and the size of the data being 500 datapoints

fig, ax = plt.subplots()
#with this command we create a figure and add an ax
ax.hist(data, bins= 200)
plt.show()
```

### Subplots

The function plt.subplots enables us to create multiple artists in a single figure. When we work with such artists, we can also add titles to our subplots and to the axes of these plots. The commands for this are a bit different than with the simple plotting functionality you were introduced to earlier. For instance, when we want to add titles to the below subplots, we can use "ax.set_title("Title") instead of plt.title("Title")

```python
data2 = np.random.normal(0, 15, 500)

fig, (ax1, ax2) = plt.subplots(2, sharex=True)
ax1.hist(data, bins= 200, color='orange')
ax2.hist(data2, bins = 200)
plt.show()

# the same is also possible in a different way
fig, axs = plt.subplots(2, sharex=True)
axs[0].hist(data, bins= 200, color='red')
axs[1].hist(data2, bins = 200)
plt.show()

# Notice how the plots are in different figures, meaning on a different canvas, each time we create a figure and show it 
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
ax1.hist(data, bins= 200, color='orange')
ax1.set_title("Data")
ax2.hist(data2, bins = 200)
ax2.set_title("Data2")
plt.show()
```

Using this functionality, we can create figures with multiple subplots in them. And we can structure these figures however we like by using the indexing functionality of python.
