```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(1000)
data2 = np.random.randn(1000)
cum_data = np.cumsum(data)
cum_data2 = np.cumsum(data2)
```

```python
def plot_data(data1, data2):
    plt.figure()
    plt.plot(cum_data)
    plt.plot(cum_data2)
    plt.xlabel('Time (ms)')
    plt.ylabel('Cumulative Random Value')
```

```python
plot_data(cum_data, cum_data2)
```

```python
import seaborn as sns
sns.set_theme()
plot_data(cum_data, cum_data2)
```
